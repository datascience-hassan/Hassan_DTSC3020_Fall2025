## 0) Setup


```python
# Run it first.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='whitegrid')
penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins.head()  # preview
```

## 1–3) Scatter plot with color by species

Create a scatter plot with:

- **x =** `bill_length_mm`
- **y =** `bill_depth_mm`

```python
# TODO: Create the scatter plot here
## 1–3) Scatter plot with color by species
# a scatter plot with:
#- **x =** `bill_length_mm`
#- **y =** `bill_depth_mm`

plt.figure(figsize=(9, 5))
sns.scatterplot(
    data=penguins,
    x='bill_length_mm',
    y='bill_depth_mm', hue='species'
)
plt.title('Bill length vs Bill depth')
plt.xlabel('Bill length (mm)')
plt.ylabel('Bill depth (mm)')
plt.show()
```

## 4) Histogram with two different bin sizes

Plot a histogram of `bill_length_mm` **twice**:

- Once with **10 bins**
- Once with **30 bins**

Then write 1–2 sentences comparing what changes when you change the number of bins.


```python
# TODO: Histogram with 10 bins
plt.figure(figsize=(9, 5))
sns.histplot(
    data=penguins,
    x='bill_length_mm',
    bins=10
)
plt.title('Distribution of bill length')
plt.xlabel('Bill length (mm)')
plt.ylabel('Count')
plt.show()

```

```python
# TODO: Histogram with 30 bins
plt.figure(figsize=(9, 5))
sns.histplot(
    data=penguins,
    x='bill_length_mm',
    bins=30
)
plt.title('Distribution of bill length')
plt.xlabel('Bill length (mm)')
plt.ylabel('Count')
plt.show()
```
